# フォルダ配下のdocxファイルを列挙する


```python
# フォルダ配下のdocxファイルを列挙する
import os
import docx
x = docx.Document() 
import re
```

```python
def get_docx_files(folder_path):
    print(folder_path)
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.endswith('.docx') and not file.startswith('~$'):
                print(os.path.join(root, file))
                yield os.path.join(root, file)
```

```python
docs = get_docx_files("../local/docs")
regex_pattern_choice = "　+[１-９,0-9]{1,2}　+"
for doc in docs:
    print(doc)
    doc = docx.Document(doc)
    for para in doc.paragraphs:
        # if paragraph text conains the pattern then print the paragraph
        if re.search(regex_pattern_choice, para.text):
            print(para.text)
    print()
```
